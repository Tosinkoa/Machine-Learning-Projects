# Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = ["Fish Pie", "Egg roll", "Donut", "Puff Puff"]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="-- Snacks --",
       ylabel="-- Total sale --")

# 5. Save & show (you save the whole figure)
fig.savefig("./sample-plot.png")
```

```python
# Make a plot from dictionary
nut_butter_prices = {"Almond butter": 10, 
                     "Peanut butter": 8, 
                     "Cashew butter": 12}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values()) # (x, y)

ax.set(title="Paul's Nut Butter",ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()))
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2

fig, ax = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot vs matplotlib OO method?)

* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method mixed with pyplot method
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age", 
            y="chol",
            c="target",
            ax=ax);

# ax.set_xlim([45, 100])
```

```python
# OO method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title ="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
over_50
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)
# Add data to ax zero
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol levels", 
       ylabel="Cholesterol")

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax one
scatter = ax1.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
# Customize ax0
ax1.set(title="Heart Disease and Cholesterol levels", 
       xlabel="Age",
       ylabel="Thalach")

# Add a legend to ax0
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")
```

```python
# Set the style
plt.style.use("seaborn-v0_8-whitegrid")

# OO method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"], 
                     y=over_50["chol"], 
                     c=over_50["target"],
                     cmap="winter") # this changes the colour skin

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title ="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--");
```

```python
# Customizing the y and x axis limitations

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2, 
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)
# Add data to ax zero
scatter = ax0.scatter(x=over_50["age"], 
                      y=over_50["chol"], 
                      c=over_50["target"],
                      cmap="winter")

# Customize ax0
ax0.set(title="Heart Disease and Cholesterol levels", 
       ylabel="Cholesterol")

# Change the x axis limit
ax0.set_xlim([50, 80])

# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="--");

# Add data to ax one
scatter = ax1.scatter(x=over_50["age"], 
                      y=over_50["thalach"], 
                      c=over_50["target"],
                      cmap="winter")
# Customize ax1
ax1.set(title="Heart Disease and Cholesterol levels", 
       xlabel="Age",
       ylabel="Thalach")

# Change ax1 x axis limits
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");

# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
